# Waarom labcontrol
Een voorzichtige conclusie na 20 jaar bachelor onderwijs is dat studenten Bsc. Elektrotechniek steeds meer moeite hebben met het correct uitvoeren van labpractica. Zeer waarschijnlijk heeft dit te maken met dalend abstract denk- en leesvermogen van de gemiddelde student. Iets dat, met de aanhoudende berichtgeving over afnemende onderwijsprestaties van Nederland op de achtergrond, aannemelijk is. Het gaat hier om het punt dat studenten zoveel moeite hebben met het lezen en nauwkeurgig uitvoeren van opdrachtbeschrijvingen, dat practica steeds vaker mislukken. Omdat metingen verkeerd uitgevoerd worden, missen studenten hun leerervaring. Die leerervaring had de ervaring moeten zijn dat abstracte theorie in de praktijk ook echt werkt. Een enorme gemiste kans, vooral vanwege de vermeende daling van abstract denkvermogen en daarmee samenhangende wiskundige vaardigheden. Een practicum bestaande uit correcte metingen met bijbehorende juiste resultaten geven de student het broodnodige plaatje van abstracte theorie. Juist dat verhelderende plaatje wordt nu niet meer bereikt. Sterker nog: studenten verlaten het lab in verwarring of zijn de weg helemaal kwijt. En dat is niet jammer, maar dat is erg. Uitgaande dat het abstract wiskundig denkniveau daalt, is een goed uitgevoerd practicum misschien wel onze enige kans om 'bewezen kennis' tussen de oren van jonge mensen te krijgen. Want telefoon en tiktok hebben niet alleen onze jongeren ook nog eens een keer opgezadeld met een enorme dosis onrust.

Het doen van meting is normaal gesproken om te bepalen of een systeem voldoet aan vooraf gestelde eisen. Tijdens een practicum, waar metingen worden uitgevoerd als onderdeel van een vak binnen een leerlijn, ligt dat iets anders. De reden om metingen door studenten te laten uitvoeren, is laten zien dat de theorie ook echt werkt. Een goed uitgevoerde meting geeft de student kans om aan de abstracte theorie een fysieke betekenis te geven. En dat is een enorme belangrijke leerervaring.

Voordat er tijdens een practicum metingen gedaan kunnen worden, zal er eerst een schakeling opgebouwd moeten worden. Daarna moet deze schakeling (Device Under Test) worden aangesloten op bijvoorbeeld labvoeding, functiegenerator en oscilloscoop:



Een meting aan een DUT wordt meestal als volt uitgevoerd:
1. De DUT wordt volgens de voorschriften opgebouwd.
2. Voedingsspanningen worden ingesteld (overeenkomstig practicumhandleiding)
3. Men herhaalt x maal een meting die er als volgt eruit ziet:
    1. Instellen signaalvorm.
    2. Instellen amplitude, offset en frequentie.
    3. Meten van de responsie van de DUT met oscilloscoop 
    4. Regristeren van de data.
4. Verwerken van alle data in een grafiek, die overeenkomt met behandelde theorie. 

## Metingen gaan vaker mis. Gemiste Kansen

Hoewel het doen van een meting niet persé moeilijk is, kan het zeker saai werk zijn. Daarnaast is het belangrijk dat labpracticanten kritisch zijn. Kritisch in en op alles :
1. Kritisch op zichzelf: een proefopstelling moeten volledig volgens instructies worden opgebouwd, alleen dan is de schakeling correct. Daarnaast geldt de regel dat de 'layout' van de opstelling altijd: compact, sterpuntsgevoed en ontkoppeld moet zijn. Dat vereist een kritische houding van diegene die de opstelling opbouwt.
2. Kritisch op de spullen: men mag er niet klakkeloos maar vanuit gaat dat de gebruikte appartuur altijd volgens verwachting functioneert,dat kabels in het lab liggen omdat ze 100% foutloos zijn of dat de voor de opstelling gebruikte componenten altijd de juiste is en altijd ook goed is en werkt.  
3. Kritisch zijn op de uitvoering practicumopdracht of -proef zelf. De regel voor practica is enorm simpel: voordat je meet moet je eerst een verwachting hebben van de uitkomst. Dat betekent dat de opdracht in zijn geheel gelezen wordt en men niet begint voordat men begrijpt waarom deze opdracht, gezien de lesstof, moet worden uitgevoerd. Een meting doen zonder te weten waarom je die opdracht doet en wat er ongeveer uit moet komen, is enorm onverstandig. Sterker nog: het is dom. Heel dom.
4. Kritisch op het resultaat. De uitkomsten moeten kloppen in het geheel. Het plaatje moet kloppen. Is dat niet zo? Vragen. Van vragen wordt je beter!

Wat ook precies de oorzaak ook is, metingen worden niet goed uitgevoerd. Daarom labcontrol, omdat je met labcontrol misschien: (a) controleren of studenten de juiste instellingen hebben gemaakt, per meetapparaat (b) of het gemetene daadwerkelijk hetgeen is dat gemeten had moeten worden of (c) studenten te helpen met de uitvoering van de meting zelf, als blijkt dat belangrijke parameters buiten specificaties van de meting dreigen te komen. De hoop daarbij is dat met labcontrol elke student de kans krijgt om met het juiste beeld of plaatje het lab te lopen waarmee de student de juiste leerervaring kan of had kunnen opdoen.
