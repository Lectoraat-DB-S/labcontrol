# Stap 3: Een connectie met de oscilloscoop opbouwen
Als het goed is draait deze Jupyter Notebookpagina binnen de Labcontrol-omgeving op jouw computer.
Benaderen van een oscilloscoop, aangesloten via USB, is dan een eitje:

```python
from devices.BaseScope import BaseScope #Om een oscilloscoop object aan te kunnen maken moet je Basescope importeren
import matplotlib.pyplot as plt         #Nodig om straks een plotje te kunnen  trekken

scope = BaseScope.getDevice()           #Hier staat: maak een BaseScope object aan en ken die tot aan variabele scope.
print(scope)                            #Druk object af: kijken of het werkt!
```

De regel waar het hier om gaat is:
`scope=BaseScope.getDevice()`

Deze regel maakt een object van het type BaseScope aan. Anders gezegd: wil je een oscilloscoop op afstand besturen, dan regel je dat door de `getDevice()` functie van de klasse `BaseScope` aan te roepen. Wil je een ander type apparaat, dan moet de getDevice() methode van de gewenste basisklasse gebruiken. De methode getDevice() van elke basisklasse zoekt binnen de Labcontrol omgeving naar een passend aangesloten device. Als die er is wordt deze aangekoppeld en in de variable `scope` opgeborgen. Via deze variabele kun je het apparaat besturen. Deze instructie gaat ervan uit dat er n oscilloscoop is aangesloten op de computer, een Siglent SDS1202 X-E oscilloscoop. In dat geval geeft `BaseScope.getDevice()` een Siglent SDS1202XE object terug. Ook dat regelt de functie `BaseScope.getDevice()` voor je.
`BaseScope.getDevice()` geeft altijd het eerste passende device terug dat ie kan vinden. Blijkt dat niet het gewenste apparaat te zijn, komt dat omdat er twee apparaten van dat type zijn aangesloten.
